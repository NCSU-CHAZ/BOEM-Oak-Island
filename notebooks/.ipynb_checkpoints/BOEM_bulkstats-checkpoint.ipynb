**Storage**

The bulk statistics are stored as .h5 files, like the other processed data files. The main difference between the bulk statistics compared to the previous data files is that the groups have been merged for the bulk statistics. This is possible since the bulk statistics are calculated only for every hour of sampling. Therefore, instead of the 14,400 samples per hour across multiple data fields, we instead are left with just one averaged value per hour for each data field.

The bulk statistics are calculated in 1 hour averages which we call 'Chunks.' Each chunk consists of several overlapping 512 second segments called ensembles that are averaged together. We choose 512 second long ensembles since the length of the ensemble determines the max period of the waves we can observe (e.g., with a shorter window, we may miss out on some longer period waves).

```python
# Import necessary packages
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import datetime
import matplotlib.dates as dt
import matplotlib.axes as axs
import matplotlib.dates as mdates
from pathlib import Path
```

```python
### Code to read in h5 files as detailed in the first guidebook

#dirpath = r'Z:\deployment_1\BulkStats\S0_103080'
dirpath=r"Z:\deployment_1\BulkStats\S0_103080" #Liz path

files = os.listdir(dirpath) # list of files in Group 1

waves = {}
for file in files:
    file_path = os.path.join(dirpath, file)  # file path
    df = pd.read_hdf(file_path)  # read the file into a Pandas dataframe
    file_name = file.split('.')[0]  # file name 
    print(file_name)
    waves[file_name] = df  # store dataframe
```

**Time**

The time data field stores the datetime assoicated with each chunk. It should be roughly evenly spaced one hour apart.

```python
print(waves['Time'])
```

```python
#Function to convert unix timestamp in ns to regular datetime
def unix_ns_to_datetime(unix_ns_array):
    dates = []
    for date in unix_ns_array:
        seconds = date // 1_000_000_000
        nanoseconds = date % 1_000_000_000
        python_datetime = datetime.datetime.utcfromtimestamp(seconds) #+ datetime.timedelta(microseconds=nanoseconds / 1000)
        dates.append(python_datetime)
    return dates
```

```python
#Applying the function to convert for datetime
unix_ns_array = waves["Time"]
waves["Datetime"] = pd.DataFrame(unix_ns_to_datetime(unix_ns_array[0]))
print(waves["Datetime"])
```

**Wave Direction and Spread**

The wave direction contains the average direction of wave groups and the spread of this measurement. The direction is measured in degrees from North, i.e. East = 90, South = 180, West = 270, North = 0 or 360. The spread refers to the deviation in degrees that the wave packets may travel in from the mean direction. So a direction of 90 and a spread of 45 means you would expect the wave packets to be traveling on average due east with parts of the wabe packet propogating from ENE to ESE.
